```python
'''
a função procurar_primos recebe dois numeros inteiros e retorna uma lista
contendo os numeros primos que estão entre os dois valores informados
'''
def procurar_primos(inicio,fim):
  lista_resultado = []
  if(inicio > fim or inicio < 0 or fim < 0):
    return False
  elif(inicio == 0 or inicio == 1):
    inicio = 2
  numero = int(inicio)
  while numero <= fim:
    i = int(2)
    primo = True
    while i < numero:
      if numero % i == 0:
        primo = False
        break
      i = i + 1
    if primo:
      lista_resultado.append(numero)
    numero = numero + 1
  return lista_resultado

# obtendo os valores
comeco = int(input('igite o comeco'))
final = int(input('digite o final'))

# fazendo o uso da função
resultado = procurar_primos(comeco,final)
```
